# Data Cleaning and Preprocessing 

## Objective
The goal of this exercise is to prepare the Car Dataset for analysis by identifying and handling missing values and outliers to ensure data quality and consistency.

### 1. Data Inspection and Summary Statistics
- **Load the Dataset:**  
  Import the Car Dataset and review its basic structure, including column names, data types, and a few initial records.

- **Generate Summary Statistics:**  
  Calculate key statistics (mean, median, min, max, standard deviation, etc.) to understand the primary characteristics of each column.

### 2. Missing Value Treatment
- **Identify Missing Values:**  
  Check for any columns with missing data and assess the extent of missingness.

- **Handle Missing Data:**  
  Replace missing values by filling them with the mean of the respective column to maintain data consistency and prevent bias.

### 3. Outlier Detection and Treatment
- **Detect Outliers:**  
  Use statistical methods, such as the Interquartile Range (IQR), to identify any extreme values that might be outliers.

- **Treat Outliers:**  
  Adjust extreme values by capping or transforming them to fall within a reasonable range. This helps create a balanced dataset, enhancing the reliability of further analysis.

### 4. Label Encoding
- **Objective**: Convert ordinal categorical data into numerical values.
- **Steps**:
  - Select a categorical column with inherent order.
  - Apply Label Encoding to transform each category into a corresponding numerical value.

### 5. One-Hot Encoding
- **Objective**: Transform non-ordinal categorical data into a format suitable for machine learning.
- **Steps**:
  - Select categorical columns without any inherent order.
  - Apply One-Hot Encoding to create binary columns for each category.



```python
# Loading all the Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```


### 1. Data Inspection and Summary Statistics

```python
df = pd.read_csv("car price.csv")
```

```python
# Review the dataset's basic structure
print("Dataset Overview:")
print(df.info()) 

print("\nFirst 5 Records:")
print(df.head()) 


print("\nSummary Statistics:")
print(df.describe()) 
```

```python
df.dtypes
```

```python
df
```

### Cleaning Columns

```python
df.rename(columns={'Engine Volume': 'Engine_Volume'}, inplace=True)
```

```python
# Convert to datetime
df['Selling_Date'] = pd.to_datetime(df['Selling_Date'], format='%d-%m-%Y', errors='coerce')  
```

```python
df['Engine_Volume'] = [col.replace('Turbo',"") for col in df['Engine_Volume']]
df['Engine_Volume'] = pd.to_numeric(df['Engine_Volume'], errors='coerce')
```

### 2. Missing Value Treatment



```python
# Identify missing values
print("Missing Values Count per Column:")
print(df.isnull().sum())  
```

Observation:-
we have missing values in kilometers_Driven and Airbags columns

```python

# Handle missing data by replacing with mean
df.fillna(df.Airbags.mean(), inplace=True) 
df.fillna(df.Kilometers_Driven.mean(), inplace=True) 
```

```python
# Verify missing values have been handled
print("\nMissing Values After Treatment:")
print(df.isnull().sum())
```

### 3. Outlier Detection and Treatment

```python
df.select_dtypes('int64','float64').head(2)
```

```python
# IQR
sns.boxplot(df['Price'])
plt.title("Box Plot before median imputation")
plt.show()
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3-q1
Lower_tail = q1 - 1.5 * iqr
Upper_tail = q3 + 1.5 * iqr
med = np.median(df['Price'])
for i in df['Price']:
    if i > Upper_tail or i < Lower_tail:
            df['Price'] = df['Price'].replace(i, med)
sns.boxplot(df['Price'])
plt.title("Box Plot after median imputation")
plt.show()   
```

```python
# Box plot before cleaning
plt.figure(figsize=(10, 6))
sns.boxplot(df['Engine_Volume'])
plt.title("Box Plot Before Column Cleaning")
plt.show()

# Cleaning: Removing rows where Engine_Volume is greater than or equal to 6
df = df[df['Engine_Volume'] < 6]

# Box plot after cleaning
plt.figure(figsize=(10, 6))
sns.boxplot(df['Engine_Volume'])
plt.title("Box Plot After Column Cleaning")
plt.show()
```

# **EXERCISE 10**

### 1. Label Encoding
- **Objective**: Convert ordinal categorical data into numerical values.
- **Steps**:
  - Select a categorical column with inherent order.
  - Apply Label Encoding to transform each category into a corresponding numerical value.

---

### 2. One-Hot Encoding
- **Objective**: Transform non-ordinal categorical data into a format suitable for machine learning.
- **Steps**:
  - Select categorical columns without any inherent order.
  - Apply One-Hot Encoding to create binary columns for each category.


## Import Libraries and Load Data


```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode categorical columns and assign back to the DataFrame
df['Manufacturer'] = label_encoder.fit_transform(df['Manufacturer'])
df['Category'] = label_encoder.fit_transform(df['Category'])
df['Leather_Interioe'] = label_encoder.fit_transform(df['Leather_Interioe'])
df['Gear_Box_Type'] = label_encoder.fit_transform(df['Gear_Box_Type'])
df['Drive_Wheel'] = label_encoder.fit_transform(df['Drive_Wheel'])
df['Wheel'] = label_encoder.fit_transform(df['Wheel'])
df['Color'] = label_encoder.fit_transform(df['Color'])
```

```python
# Columns for One-Hot Encoding
columns_to_one_hot_encode = ['Model', 'Drive_Wheel','Leather_Interioe','Fuel_Type','Category']

# Apply One-Hot Encoding
df_one_hot = pd.get_dummies(df, columns=columns_to_one_hot_encode, drop_first=True, dtype=int)
```

```python
# Combine both label-encoded and one-hot encoded data
df_combined = pd.concat([df, df_one_hot], axis=1)

# Drop original categorical columns if needed
df_combined.drop(columns=columns_to_one_hot_encode, inplace=True)

```
