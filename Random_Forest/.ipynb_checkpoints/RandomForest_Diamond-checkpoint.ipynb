# Random Forest Classifier Exercise

## Objective  
The goal of this exercise is to build a **Random Forest Classifier** to predict whether an individual has heart disease (`target`) using the provided dataset.

---

## 1. Data Dictionary  

**Dataset Link**: *[Heart Disease Classifier](https://drive.google.com/file/d/1LNRp58nqM0CHkHyKV9YkDVMFsUYV2cmv/view?usp=sharing)*

| **Feature**   | **Description**                                                          |
|---------------|--------------------------------------------------------------------------|
| `age`        | Age of the person (numerical).                                            |
| `sex`        | Gender (0 = female, 1 = male).                                            |
| `cp`         | Chest pain type (0–3 categorical).                                        |
| `trestbps`   | Resting blood pressure in mm Hg (numerical).                              |
| `chol`       | Serum cholesterol in mg/dl (numerical).                                   |
| `fbs`        | Fasting blood sugar > 120 mg/dl (0 = false, 1 = true).                    |
| `restecg`    | Resting electrocardiographic results (0 = normal, 1 = abnormal, 2 = LVH). |
| `thalach`    | Maximum heart rate achieved (numerical).                                  |
| `exang`      | Exercise-induced angina (0 = no, 1 = yes).                                |
| `oldpeak`    | ST depression induced by exercise relative to rest (numerical).           |
| `slope`      | The slope of the peak exercise ST segment (0–2 categorical).              |
| `ca`         | Number of major vessels (0–3) colored by fluoroscopy (numerical).         |
| `thal`       | Thalassemia (0 = normal, 1 = fixed defect, 2 = reversible defect, 3).     |
| `target`     | Presence of heart disease (0 = no, 1 = yes).                              |

---

Build a **Random Forest Classifier** to predict the target (presence of heart disease) based on the features provided.  
The aim is to create an **accurate** and **interpretable** model.

## 4. Model Building  

```python
X = df.drop("target",axis=1)
y =df[['target']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.20,random_state=40)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

**Random Forest Classifier with Hyperparameter Tuning**

```python
# Define hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 150],           
    'max_depth': [None, 10, 20, 30],          
    'min_samples_split': [2, 5, 10],          
    'min_samples_leaf': [1, 2, 4],            
    'criterion': ['gini', 'entropy']         
}
```

```python
# Initialize the RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=42)
```

```python
# Train the model with GridSearchCV
rf_clf.fit(X_train, y_train)
```

```python
# Best Hyperparameters
# print("Best Hyperparameters:", rf_clf.best_params_)
```

```python
# Train the best model
# best_rf = rf_clf.best_estimator_
```

**Make Predictions** on the test set

```python
y_pred = rf_clf.predict(X_test)
print(y_pred)
```

### **Model Evaluation**

```python
# Accuracy Score and Precision Score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
precision = precision_score(y_test,y_pred)
print("Precision Score:",precision)
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python

# Plot Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
