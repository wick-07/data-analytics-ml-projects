# *Cleaning Data*

```python
car_dataset.rename(columns={"Engine Volume":"Engine_Volume"},inplace=True)
```

```python
car_dataset.head()
```

```python
car_dataset["Engine_Volume"]=[col.replace("Turbo","") for col in car_dataset["Engine_Volume"]]
```

```python
car_dataset.head()
```

```python
car_dataset["Engine_Volume"].dtype
```

```python
car_dataset["Engine_Volume"]= pd.to_numeric(car_dataset["Engine_Volume"],errors="coerce")
```

```python
car_dataset["Engine_Volume"].dtype
```

```python
car_dataset["Prod_Year"].dtype
```

```python
car_dataset["Prod_Year"]=pd.to_datetime(car_dataset["Prod_Year"],format="%Y",errors="coerce").dt.year
```

```python
car_dataset["Prod_Year"].dtype
```

```python
car_dataset["Selling_Date"].dtype
```

```python
car_dataset["Selling_Date"]
```

```python
car_dataset['Selling_Date']=pd.to_datetime(car_dataset['Selling_Date'],format='%d-%m-%Y',errors='coerce')
```

```python
car_dataset['Selling_Date'].dtype
```

# *Handling Missing Values*

```python
car_dataset.isnull().sum()
```

```python
car_dataset["Kilometers_Driven"]=car_dataset["Kilometers_Driven"].fillna(car_dataset["Kilometers_Driven"].mean())
car_dataset.isnull().sum()
```

```python
car_dataset["Airbags"]=car_dataset["Airbags"].fillna(car_dataset["Airbags"].mean())
car_dataset.isnull().sum()
```

# *Outliers Treatment*

```python
car_dataset.select_dtypes('int64','float64').head(2)
```

```python
car_dataset.select_dtypes(['int64','float64']).head(2)
```

```python
plt.boxplot(car_dataset["Price"])
plt.title("Boxplot of price before treating outliers")
plt.show()
q1=car_dataset["Price"].quantile(0.25)
q3=car_dataset["Price"].quantile(0.75)
iqr=q3-q1
lower_tail=q1-1.5*iqr
upper_tail=q3+1.5*iqr
med=car_dataset["Price"].median()

for i in car_dataset["Price"]:
    if i<lower_tail or i>upper_tail:
        car_dataset["Price"]=car_dataset["Price"].replace(i,med)
sns.boxplot(car_dataset["Price"])        
plt.title("Boxplot of Price after treating outliers")
plt.show()
```

# *Bivariate and Multivariate Analysis*

```python
Manufacturer_ByPrice=car_dataset.groupby(car_dataset["Manufacturer"])["Price"].max().reset_index().sort_values(by='Price',ascending=False)
display(Manufacturer_ByPrice)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=Manufacturer_ByPrice, x='Manufacturer', y='Price',hue='Manufacturer',palette='viridis',legend=False)
plt.title('Maximum Price by Manufacturer')
plt.xticks(rotation=85)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=car_dataset, x='Drive_Wheel', y='Price')
plt.title('Price Distribution by Drive Wheel')
plt.show()
```

```python
df_pivot = car_dataset.groupby(['Manufacturer', 'Category'])['Price'].max().unstack().fillna(0)

# Plotting the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(df_pivot, annot=True, fmt='g', cmap='coolwarm', linewidths=0.5)
plt.title('Price by Manufacturer and Category (Heatmap)')
plt.xlabel('Category')
plt.ylabel('Manufacturer')
plt.show()
```
